```python
mrna_df =pd.read_csv('/tf/Documents/dsi/capstone2/capstone2/data/capstone2.mrn_df2.csv')
df = pd.read_csv('/tf/Documents/dsi/capstone2/capstone2/data/METABRIC_RNA_Mutation.csv')
```

```python
death_from_dict = {
    'Living':0
    ,'Died of Other Causes':0
    ,'Died of Disease':1
}
df.replace(death_from_dict, inplace =True)
mrna_df['death_from_cancer'] = df.death_from_cancer
```

```python
mrna_df.death_from_cancer.fillna(0, inplace = True)
```

```python
y = mrna_df.pop('death_from_cancer')
```

```python
X = mrna_df
```

```python
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.15, random_state=1, stratify = y)

X_train2, X_test, y_train2, y_test = train_test_split(X_train.copy(), y_train.copy(), test_size=0.10, random_state=1)
```

```python
NN_model = keras.Sequential([
    layers.Dense(32, activation='relu',input_shape=X_train2[1].shape),
    layers.Dense(16, activation='relu'),
    layers.Dense(1, activation='sigmoid')
])
```

```python
NN_model.compile(
    loss='binary_crossentropy',
    optimizer='adam',
    metrics=['binary_accuracy']
)
```

```python
early_stopping = keras.callbacks.EarlyStopping(
    patience=5,
    min_delta=0.005,
    restore_best_weights=True,
)
```

```python
train_history = NN_model.fit(
    X_train2, y_train2,
    validation_data=(X_test, y_test),
    batch_size = 8,
    epochs = 200,
     callbacks=[early_stopping]
)
```

```python
model = Sequential()
model.add(Dense(12, input_dim=489, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(11, activation='relu'))
model.add(Dropout(0.20))
model.add(Dense(1, activation='relu'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(X_train2, y_train2, epochs=160, batch_size=10)
```

```python
_, accuracy = model.evaluate(X_test, y_test)
print('Accuracy: %.2f' % (accuracy*100))
```

```python
_, accuracy = model.evaluate(X_val, y_val)
```
